# **Analysis of House Price Prediction using Regression models**

This document presents an analysis of a dataset related to house pricing, with the primary objective of implementing several regression models. These selected models have been chosen for their efficacy in managing multicollinearity and offering regularized solutions, making them particularly suitable for conducting in-depth analysis in the real estate market.

$\underline{\text{Regression Models}}:$

- $\textbf{Linear Regression:}$

  Linear regression is a fundamental model used to establish a linear relationship between the dependent variable (house price) and one or more independent variables (predictors). It provides insights into the impact of each predictor on house prices.

- $\textbf{Polynomial Regression:}$

  Polynomial regression extends linear regression by allowing for nonlinear relationships between predictors and the target variable. This model can capture more complex patterns in the data.

- $\textbf{Ridge Regression:}$

  Ridge regression is a regularization technique that mitigates multicollinearity by adding a penalty term to the linear regression cost function. This helps prevent overfitting and provides more stable coefficient estimates.

- $\textbf{Lasso Regression:}$

  Lasso regression, similar to Ridge regression, introduces regularization to linear regression. However, it employs L1 regularization, which tends to produce sparse coefficient estimates by setting some coefficients to exactly zero. This feature selection property can be valuable in feature engineering.

- $\textbf{Random Forest Regression:}$

  Random forest is an ensemble learning method that combines multiple decision trees to make predictions. In regression tasks, it can capture complex interactions among variables and provide robust predictions.

- $\textbf{Support Vector Machines (SVM) Regression:}$

  SVM regression is a powerful algorithm for modeling nonlinear relationships. It identifies a hyperplane that best fits the data while maximizing the margin between data points and the hyperplane, resulting in robust predictions.

# Data Analysis

  To calculate the correlation between two variables, we will use a statistical measure called the correlation coefficient. The most common correlation coefficient used is the Pearson correlation coefficient (also known as Pearson's Correlation Coefficient - PCC).

  In the following, we define how to calculate the Pearson correlation coefficient between 2 variables or features X and Y:

1. Calculate the means: We compute the mean of both X and Y values, denoted as $\bar{X}$ and $\bar{Y}$.

2. Calculate the covariance: We compute the covariance between X and Y values.
  
  The formula for covariance is:

  $$ Cov(X, Y) = \sum_{i=1}^{n} \frac{(X_{i} - \bar{X}) \cdot (Y_{i} - \bar{Y})}{n - 1} $$

  Where:
  - Cov(X, Y) is the covariance between X and Y.
  - $X_{i}$ and $Y_{i}$ are individual data points of X and Y.
  - $\bar{X}$ is the mean of X.
  - $\bar{Y}$ is the mean of Y.
  - $n$ is the number of data points.

3. Calculate the standard deviations: We ompute the standard deviation of both X and Y.

  The formula for standard deviation is:

  $$ SD(X) = \sqrt{\sum_{i=1}^{n} \frac{(X_{i} - \bar{X})^2}{n - 1}} $$
  $$ SD(Y) = \sqrt{\sum_{i=1}^{n} \frac{(Y_{i} - \bar{Y})^2}{n - 1}} $$

  Where:
  - $SD(X)$ is the standard deviation of height.
  - $SD(Y)$ is the standard deviation of weight.
  - $X_{i}$ and $Y_{i}$ are individual data points of height and weight.
  - $\bar{X}$ is the mean of height.
  - $\bar{Y}$ is the mean of weight.
  - $n$ is the number of data points.

4. Calculate the Pearson Correlation Coefficient:

    Finally, we can calculate the Pearson Correlation Coefficient using the following formula:

  $$ r = \frac{Cov(X, Y)}{SD(X) \cdot SD(Y)} $$

5. Interpretation of values:

    The resulting value of r will be between -1 and 1.
    - If r is close to 1, it indicates a strong positive linear correlation (as one variable increases, the other tends to increase).
    - If r is close to -1, it indicates a strong negative linear correlation (as one variable increases, the other tends to decrease and viceversa).
    - If r is close to 0, it suggests a weak or no linear correlation between the two variables.

## 1. Linear Regression
Building First model of regression - Linear Regression

```python
# Create a pipeline that first standardizes the data then applies Linear regression
# StandardScaler will normalize the features (subtract mean and divide by standard-deviation)
pipeline_linear = Pipeline([
    ('scaler', StandardScaler()),
    ('linear', LinearRegression())
])

param_grid = {}

# Initialize GridSearchCV
grid_linear = GridSearchCV(pipeline_linear, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the model on the training data
grid_linear.fit(X_train, Y_train)

# The best model from grid search
best_linear_model = grid_linear.best_estimator_
```

In order to evaluate the model's prediction, we will consider 2 main metrics.

$\underline{\text{Definition of the metrics}}$:

1. RMSE (Root Mean Square Error)

$$ RMSE ={\sqrt {{\frac{1}{n}}\sum _{i=1}^{n}(X_{i}-\hat{X_{i}})^{2}}} $$

where
$n$ is the number of observations, $X_{i}$ is the actual observed value and $\hat{X_{i}}$ is the predicted value from the model.

$\underline{Interpretation}:$

- An RMSE of 0 indicates that the model makes perfect predictions with no errors, which is typically not possible with real-world data.

- The RMSE is always non-negative, and a lower value indicates a better fit. The closer the RMSE is to 0, the better the model is at predicting the observed outcomes.

2. R2-Score

$$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{\sum_{i=1}^{n} (X_i - \hat{X}_i)^2}{\sum_{i=1}^{n} (X_{i} - \bar{X})^2}$$

The terms $SS_{res}$(Sum of Squares of Residuals) and $ SS_{tot}$(Total Sum of Squarees) are defined as follows:

$$ SS_{res} = \sum_{i=1}^{n} (X_i - \hat{X}_i)^2 $$

$$ SS_{tot} = \sum_{i=1}^{n} (X_{i} - \bar{X})^2 $$

Where $\bar{X}$ denotes the mean of X.

$\underline{Interpretation}:$

- An R² score of 1 indicates that the regression predictions perfectly fit the data.
- An R² score of 0 indicates that the model explains none of the variability of the response data around its mean.
- Negative values of R² may occur when the model fits the data worse than a horizontal hyperplane at the mean of the dependent variable.

```python
# Define the Real & Prediction Values
Y_prediction_linear_test = best_linear_model.predict(X_test)
Y_prediction_linear_train = best_linear_model.predict(X_train)

# Evaluate the model with the chosen metrics on test set
RMSE_score_linear_test = np.sqrt(mean_squared_error(Y_test, Y_prediction_linear_test))
R2_score_linear_test = r2_score(Y_test, Y_prediction_linear_test)

# Evaluate the model with the chosen metrics on training set
RMSE_score_linear_train = np.sqrt(mean_squared_error(Y_train, Y_prediction_linear_train))
R2_score_linear_train = r2_score(Y_train, Y_prediction_linear_train)

print(f'Test RMSE: {RMSE_score_linear_test}, Test R2: {R2_score_linear_test}')
print(f'Train RMSE: {RMSE_score_linear_train}, Train R2: {R2_score_linear_train}')
```

## 2. Polynomial Regression
Building Second model of regression - Polynomial Regression

```python
# Create a pipeline that first standardizes the data, applies PolynomialFeatures, and then applies Linear Regression
# StandardScaler will normalize the features (subtract mean and divide by standard-deviation)
# PolynomialFeatures will create polynomial features up to the specified degree
# Linear Regression will perform regression on the polynomial features

pipeline_poly = Pipeline([
    ('scaler', StandardScaler()),
    ('poly', PolynomialFeatures()),
    ('linear', LinearRegression())
])

# Parameters for GridSearchCV
# Here, we are trying different polynomial degrees for Polynomial regression (if needed)
param_grid = {
    'poly__degree': [2, 3, 4],
}

# Initialize GridSearchCV
grid_polynomial = GridSearchCV(pipeline_poly, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the model on the training data
grid_polynomial.fit(X_train, Y_train)

# The best model from grid search
best_polynomial_model = grid_polynomial.best_estimator_

# Print the best model parameters
print(f'Best parameters for Polynomial Regression: {grid_polynomial.best_params_}')
```

```python
# Define the Real & Prediction Values
Y_prediction_polynomial_test = best_polynomial_model.predict(X_test)
Y_prediction_polynomial_train = best_polynomial_model.predict(X_train)

# Evaluate the model with the chosen metrics on test set
RMSE_score_polynomial_test = np.sqrt(mean_squared_error(Y_test, Y_prediction_polynomial_test))
R2_score_polynomial_test = r2_score(Y_test, Y_prediction_polynomial_test)

# Evaluate the model with the chosen metrics on training set
RMSE_score_polynomial_train = np.sqrt(mean_squared_error(Y_train, Y_prediction_polynomial_train))
R2_score_polynomial_train = r2_score(Y_train, Y_prediction_polynomial_train)

print(f'Test RMSE: {RMSE_score_polynomial_test}, Test R2: {R2_score_polynomial_test}')
print(f'Train RMSE: {RMSE_score_polynomial_train}, Train R2: {R2_score_polynomial_train}')
```

## 3. Ridge Regression
Building Third model of regression - Ridge Regression

## 5. Random forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

# define Pipeline
# This code defines a Pipeline object with two steps: first, it uses StandardScaler to standardize the features
# by removing the mean and scaling to unit variance, and second, it applies a RandomForestRegressor for regression.
pipeline_rf = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestRegressor(random_state=123))
])

# Setting Up the Parameter Grid
# This is the parameter grid for the grid search. It specifies the parameters to be tested: n_estimators (the number of trees in the forest),
# max_features (the number of features to consider when looking for the best split), and max_depth (the maximum depth of the trees).
# I limit the depth of the tree to prevent overfitting
# Adjust the number of samples for the minimum node
param_grid_rf = {
    'rf__n_estimators': [100, 200],
    'rf__max_features': ['log2', 'sqrt'],
    'rf__max_depth': [10, 20, None]
}

# initialisation GridSearchCV
grid_rf = GridSearchCV(pipeline_rf, param_grid_rf, cv=5, scoring='neg_mean_squared_error')

# train model
grid_rf.fit(X_train, Y_train)

# get a best model
best_rf_model = grid_rf.best_estimator_
```

```python
# Define the Real & Prediction Values
Y_prediction_rf_test = best_rf_model.predict(X_test)
Y_prediction_rf_train = best_rf_model.predict(X_train)
# Evaluate the model with the chosen metrics on test set
RMSE_score_rf_test = np.sqrt(mean_squared_error(Y_test, Y_prediction_rf_test))
R2_score_rf_test = r2_score(Y_test, Y_prediction_rf_test)
# Evaluate the model with the chosen metrics on training set
RMSE_score_rf_train = np.sqrt(mean_squared_error(Y_train, Y_prediction_rf_train))
R2_score_rf_train = r2_score(Y_train, Y_prediction_rf_train)
print("Random forest Train score",RMSE_score_rf_train, R2_score_rf_train)
print("Random forest Test score",RMSE_score_rf_test,R2_score_rf_test)
```

```python
feature_names = df_housing.drop('SalePrice', axis=1).columns
# Get the feature importance of the random forest model
feature_importances = best_rf_model.named_steps['rf'].feature_importances_

# Create a feature importance map
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances, y=feature_names)
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importances')
plt.show()
```

```python
# prepare a list for stock des resultsets
results = []

# loop differents features
for num_features in range(1, len(feature_importances) + 1):
    # Get top features index and get features
    indices = np.argsort(feature_importances)[-num_features:]
    top_features = feature_names[indices]

    # Recreate training and test sets
    X_train_reduced = X_train[:, indices]
    X_test_reduced = X_test[:, indices]

    # Retrain the model
    rf_reduced = RandomForestRegressor(n_estimators=best_rf_model.named_steps['rf'].n_estimators,
                                       max_features=best_rf_model.named_steps['rf'].max_features,
                                       max_depth=best_rf_model.named_steps['rf'].max_depth,
                                       random_state=123)
    rf_reduced.fit(X_train_reduced, Y_train)

    # make predictions
    Y_prediction_reduced = rf_reduced.predict(X_test_reduced)
    Y_prediction_reduced_train = rf_reduced.predict(X_train_reduced)
    # Evaluate and compare the performance of new models
    RMSE_score_reduced = np.sqrt(mean_squared_error(Y_test, Y_prediction_reduced))
    R2_score_reduced = r2_score(Y_test, Y_prediction_reduced)
    # Evaluate the model with the chosen metrics on training set of new models
    RMSE_score_rf_train = np.sqrt(mean_squared_error(Y_train, Y_prediction_reduced_train))
    R2_score_rf_train = r2_score(Y_train, Y_prediction_reduced_train)
    print("Random forest Train score",RMSE_score_rf_train, R2_score_rf_train)
    # Save results to list
    results.append((num_features, RMSE_score_reduced, R2_score_reduced))

# print all the results
for result in results:
    print(f"Number of Features: {result[0]}, RMSE: {result[1]}, R2 Score: {result[2]}")
# After filtering feature importance and comparing various accuracy rates,
# I found that  need to remove the last one feature, the R2_score will performe better.
num_features_list, RMSE_score_reduced, R2_score_reduced = zip(*results)
plt.figure(figsize=(10, 6))
plt.plot(num_features_list,R2_score_reduced, marker='o', label='Test R2 Score')
plt.xlabel("Numbre of Features")
plt.ylabel("R2 Score")
plt.title('R2 Score by Number of Features Used')
plt.legend()
plt.show()
```
